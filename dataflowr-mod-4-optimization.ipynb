```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_trace_2d(f, res, color='red'):
    x1,x2 = zip(*res)
    plt.plot(x1,x2, '-o', color=color)
    x1 = np.arange(-5.5, 3.5, 0.1)
    x2 = np.arange(min(-3.0, min(x2)-1), max(1.0, max(x2)+1), 0.1)
    x1, x2 = np.meshgrid(x1, x2)
    plt.contour(x1, x2, f(x1, x2), colors='blue')
    plt.xlabel('x1')
    plt.ylabel('x2')
    
    
```

```python
def train_2d(trainer):
    x1, x2 = -5, -2
    s_x1, s_x2 = 0,0
    res=[(x1, x2)]
    for i in range(20):
        x1,x2, s_x1, s_x2 = trainer(x1, x2, s_x1, s_x2)
        res.append((x1, x2))
    print('epoch %d , x1 %f, x2 %f' % (i+1,x1, x2))
    return res
```

***Gradient Descent Algorithm***

```python
## Constant eta aka learning rate
eta = 0.4
def f_2d(x1, x2):
    return 0.1 * x1**2 + 2 * x2**2

def gd_2d(x1,x2, s1 ,s2):
    return (x1 - eta * 0.2 * x1, x2 - eta*4*x2, 0,0)
```

```python
show_trace_2d(f_2d, train_2d(gd_2d))
```

```python
## for learning rate 0.4
eta = 0.6
show_trace_2d(f_2d, train_2d(gd_2d))
```

***Momentum***

```python
eta, gamma = 0.4, 0.5 
def momentum_2d(x1, x2, v1, v2):
    v1 = gamma * v1 + eta * x1 * 0.2
    v2 = gamma * v2 + eta * x2 * 4
     
    return x1-v1, x2-v2, v1, v2
```

```python
show_trace_2d(f_2d, train_2d(momentum_2d))
```

```python
eta = 0.6
show_trace_2d(f_2d, train_2d(momentum_2d))
```

```python
eta, gamma = 0.05, 0.9
show_trace_2d(f_2d, train_2d(momentum_2d))
```

***Nesterov Accelerated Gradients***


```python
eta, gamma = 0.05, 0.9
def nesterov_2d(x1, x2, v1, v2):
    v1 = gamma * v1 + eta * 0.2  * (x1 - gamma * v1)
    v2 = gamma * v2 + eta * 4  * (x2 - gamma * v2)
    return x1 - v1, x2 - v2, v1, v2
```

```python
show_trace_2d(f_2d, train_2d(nesterov_2d))
show_trace_2d(f_2d, train_2d(momentum_2d), color= 'green')
```

***AdaGrad***

```python
import math

def adagrad_2d(x1, x2, s1, s2):
    g1,g2, eps = 0.2 * x1, 4 * x2, 1e-6
    
    s1 = s1 + g1 ** 2 
    s2 = s2 + g2 ** 2 
    
    x1 = x1 - (eta/(math.sqrt(s1+eps))) * g1
    
    x2 = x2 - (eta/(math.sqrt(s2+eps))) * g2
    
    
    return x1, x2 , s1, s2
```

```python
eta = 0.4 
show_trace_2d(f_2d, train_2d(adagrad_2d))
```

```python
eta = 2
show_trace_2d(f_2d, train_2d(adagrad_2d))
```

***RMSProp***

```python
def rmsprop_2d(x1, x2, s1, s2):
    g1, g2 , eps = 0.2 * x1, 4 * x2, 1e-6
    
    s1 = gamma * s1 + (1 - gamma) * g1 ** 2
    s2 = gamma * s2 + (1 - gamma) * g2 ** 2
    x1 -= (eta/(math.sqrt(s1 + eps))) * g1
    x2 -= (eta/(math.sqrt(s2 + eps))) * g2
    
    return x1, x2, s1, s2
```

```python
eta , gamma = 0.4, 0.9
show_trace_2d(f_2d, train_2d(rmsprop_2d))
```

***ADAM Optimizer***

```python
def adam_2d(x1, x2, s1, s2, t):
    beta1, beta2, eps = 0.9, 0.99, 1e-6
    g1, g2 = 0.2 * x1, 4 * x2
    s1[0] = beta1 * s1[0] + (1 - beta1) * g1
    s2[0] = beta1 * s2[0] + (1 - beta1) * g2
    s1[1] = beta2 * s1[1] + (1 - beta2) * g1 ** 2
    s2[1] = beta2 * s2[1] + (1 - beta2) * g2 ** 2
    
    m1 = s1[0]/(1 - beta1 ** (t+1))
    m2 = s2[0]/(1 - beta1 ** (t+1))
    v1 = s1[1]/(1 - beta2 ** (t+1))
    v2 = s2[1]/(1 - beta2 ** (t+1))
    x1 -= eta / (math.sqrt(v1) + eps) * m1
    x2 -= eta / (math.sqrt(v2) + eps) * m2
    
    return x1 , x2, s1, s2
    
    
```

```python
def train_adam_2d(trainer):
    x1, x2 = -5, -2
    s_x1 , s_x2 = [0,0,0], [0,0,0]
    
    res = [(x1, x2)]
    
    for i in range(20):
        x1 , x2 , s_x1, s_x2 = trainer(x1, x2, s_x1, s_x2, i)
        res.append((x1,x2))
    
    print('epoch %d, x1 %f, x2 %f' % (i+1, x1, x2))
    
    return res
        
```

```python
def amsgrad_2d(x1, x2, s1, s2 , t):
    beta1, beta2, eps = 0.9, 0.99, 1e-6
    g1, g2 = 0.2 * x1, 4 * x2
    g1, g2 = 0.2 * x1, 4 * x2
    s1[0] = beta1 * s1[0] + (1 - beta1) * g1
    s2[0] = beta1 * s2[0] + (1 - beta1) * g2
    s1[1] = beta2 * s1[1] + (1 - beta2) * g1 ** 2
    s2[1] = beta2 * s2[1] + (1 - beta2) * g2 ** 2
    m1 = s1[0]#/(1-beta1 ** (t+1))
    m2 = s2[0]#/(1-beta1 ** (t+1))
    v1 = max(s1[2], s1[1])
    v2 = max(s2[2],s2[1])
    x1 -= eta / (math.sqrt(v1) + eps) * m1
    x2 -= eta / (math.sqrt(v2) + eps) * m2
    s1[2] = v1
    s2[2] = v2
    return x1, x2, s1, s2
```

```python
eta = 1
show_trace_2d(f_2d, train_adam_2d(amsgrad_2d))
show_trace_2d(f_2d, train_adam_2d(adam_2d), color ='green')
```
