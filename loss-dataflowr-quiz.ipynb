```python
import torch 
import torch.nn as nn
```

```python
m = nn.Sigmoid()
loss = nn.BCELoss()
target = torch.empty(3).random_(2)
print(m, loss,target)
```

```python
input = torch.randn(3, requires_grad = True)
optimizer = torch.optim.SGD([input], lr = 0.1)
print(optimizer)
print(input,m(input))
```

```python
for _ in range(10000):
    output = loss(m(input), target)
    optimizer.zero_grad()
    output.backward()
    optimizer.step()

# print(m(target))
```

```python
print(input, m(input), target,output, output.item())
```

```python
loss2 = nn.BCEWithLogitsLoss()
target = torch.empty(3).random_(2)
print(target)

input = torch.randn(3, requires_grad=True)
optimizer = torch.optim.SGD([input], lr = 0.1)
for _ in range(500):
    output = loss2(m(input), target)
    optimizer.zero_grad()
    output.backward()
    optimizer.step()
print(m(input))
```

```python
loss3 = nn.NLLLoss()
m3 = nn.LogSoftmax(dim=1)
target = torch.empty(4).random_(6)
print(target, target.size())

print(loss3, m3)
```

```python
input = torch.randn(4, requires_grad=True)
print(input.size())
optimizer = torch.optim.SGD([input], lr = 0.1)
for _ in range(1000):
    output = loss3(m3(input), target)
    optimizer.zero_grad()
    output.backward()
    optimizer.step()
print(torch.exp(m3(input)))
```

```python
input = torch.randn((4,7), requires_grad=True)
optimizer = torch.optim.SGD([input], lr = 0.1)
for _ in range(1000):
    output = loss3(m3(input), target.long())
    optimizer.zero_grad()
    output.backward()
    optimizer.step()
print(torch.exp(m3(input)))
```
